```python
import pandas as pd
from docx import Document

path = r'../CONTRATO ADITIVO PADRÃO.docx'

# dicionário para modificar
dict_modification = {
    "@num_aditivo@" : "1",
    "@razao_social@" : "Razão Social Exemplo",
    "@cnpj@" : "00.000.000/0001-00",
    "@cidade@" : "Cidade Exemplo",
    "@endereco@" : "Endereco Exemplo",
    "@numero@" : "1234567890",
    "@bairro@" : "Bairro Exemplo",
    "data_contrato@" : "01 de Agosto de 2025",
    "@data_retroagem@" : "02 de Setembro de 2025",
    "@valor@" : "1.000,00",
    "@valor_descrito@" : "Um mil reais",
    "@nome_contratado@" : "NOME TESTE"
}
list_modification = []

doc = Document(path)

for para in doc.paragraphs:
    for run in para.runs:
        for key, value in dict_modification.items():
            if key in run.text:
                run.text = run.text.replace(key, value)
                lista_de_substituicao = f"Substituído '{key}' por '{value}'"
                list_modification.append(lista_de_substituicao)
    print(para.text)

print("\n")
print(100 * '-')
for item in list_modification:
    print(item)

doc.save(r'../Arquivos/CONTRATO ADITIVO MODIFICADO.docx')
```

```python
import pandas as pd
from num2words import num2words

def value_title(serie):
    return serie.str.title()
# Lista dos valores
valores_str = [
    "R$ 144.000,00", "R$ 84.000,00", "R$ 180.000,00",
]

# Convertendo os valores para float
valores_float = [float(v.replace("R$ ", "").replace(".", "").replace(",", ".")) for v in valores_str]

# Convertendo para texto por extenso
valores_por_extenso = [num2words(v, lang='pt_BR', to='currency') for v in valores_float]
    
# Criando DataFrame
df = pd.DataFrame({
    "valor": valores_str,
    "descrição": valores_por_extenso
})
df['descrição'] = value_title(df['descrição'])
# Exibindo os primeiros resultados

for i in range(len(df)):
    print(f"Valor: {df['valor'][i]} {df['descrição'][i]}")
```

## Script para criar os contratos de Aditivos

```python
import pandas as pd
import locale
from num2words import num2words
from datetime import datetime
from docx import Document


locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')

def value_descrition(serie):
    valores_float = [float(v.replace("R$ ", "").replace(".", "").replace(",", ".")) for v in serie]
    serie = [num2words(v, lang='pt_BR', to='currency') for v in valores_float]
    return serie

def value_title(serie):
    return serie.str.capitalize()


def value_money(valor):
    locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
    valor_formatado = locale.format_string('%.2f', valor, grouping=True)
    return valor_formatado

def data_por_extenso(data_str):
    # Converte a string em um objeto datetime
    data = datetime.strptime(data_str, "%d/%m/%Y")
    
    # Lista com os nomes dos meses
    meses = [
        "janeiro", "fevereiro", "março", "abril", "maio", "junho",
        "julho", "agosto", "setembro", "outubro", "novembro", "dezembro"
    ]
    
    # Monta a data por extenso
    return f"{data.day} de {meses[data.month - 1]} de {data.year}"

path = r'../CONTRATO ADITIVO PADRÃO.docx'
path_csv = r'../modelo_csv.csv'

df = pd.read_csv(path_csv, sep=';', encoding='latin1')

df['NUM_ADITIVO'] = df['NUM_ADITIVO'].astype(str)

df['VALOR_DESCRICAO'] = value_descrition(df['NOVO_VALOR'])
df['VALOR_DESCRICAO'] = value_title(df['VALOR_DESCRICAO'])

# Converta NOVO_VALOR para float antes de aplicar value_money
df['NOVO_VALOR'] = df['NOVO_VALOR'].apply(lambda x: float(str(x).replace("R$ ", "").replace(".", "").replace(",", ".")))
df['NOVO_VALOR'] = df['NOVO_VALOR'].apply(value_money)
df['NOVO_VALOR'] = 'R$ ' + df['NOVO_VALOR']

# datas agora
df['DATA_CONTRATO_EXTENSO'] = df['DATA_CONTRATO'].apply(data_por_extenso)
df['DATA_RETROAGE_EXTENSO'] = df['DATA_RETROAGE'].apply(data_por_extenso)

print(df.columns)
df.head()

name_mes = "AGOSTO"

for num, nome in enumerate(df['NOME']):
    dict_modification = {
    "@num_aditivo@" : df['NUM_ADITIVO'].iloc[num],
    "@razao_social@" : df['RAZAO_SOCIAL'].iloc[num],
    "@cnpj@" : df['CNPJ'].iloc[num],
    "@cidade@" : df['CIDADE'].iloc[num],
    "@endereco@" : df['ENDERECO'].iloc[num],
    "@numero@" : df['NUMERO'].iloc[num],
    "@bairro@" : df['BAIRRO'].iloc[num],
    "data_contrato@" : df['DATA_CONTRATO_EXTENSO'].iloc[num],
    "@data_retroagem@" : df['DATA_RETROAGE_EXTENSO'].iloc[num],
    "@valor@" : df['NOVO_VALOR'].iloc[num],
    "@valor_descrito@" : df['VALOR_DESCRICAO'].iloc[num],
    }

    doc = Document(path)

    for para in doc.paragraphs:
        for run in para.runs:
            for key, value in dict_modification.items():
                if key in run.text:
                    run.text = run.text.replace(key, str(value))
                    name_arquivo = f'CONTRATO ADITIVO - {nome} - {name_mes}.docx'
                    doc.save(name_arquivo)

    print(nome)
    print(dict_modification)
```
