## Script para recisão com aviso

```python
import pandas as pd
from docx import Document
from docx2pdf import convert
import os

path_recisao = r'../Arquivos/RECISÃO CONTRATUAL - HAP - NOTIFICAÇÃO COM AVISO.docx'
path = r'../modelo_recisao.csv'
df = pd.read_csv(path, sep=';', encoding='latin1')

# for num, nome in enumerate(df['NOME']):
#     #dicionário para modificar
#     dict_modification = {
#         "[DATA_DESLIGAMENTO]" : "",
#         "@razao_social@" : "Razão Social Exemplo",
#     }
#     list_modification = []

#dicionário para modificar
dict_modification = {
    "[DATA_DESLIGAMENTO]" : "01/01/2020",
    "[NOME_CONTRATADA]" : "FALUNO DE TAL",
    "[CNPJ_CONTRATADA]": "00.000.000/0001-00",
    "[DATA_CONTRATACAO]":"0/01/2020",
    "[DIAS_EFEITO]": "30 (trinta dias)",
    "@dia@": "30 (trinta dias)",
    }

list_modification = []

doc = Document(path_recisao)

for para in doc.paragraphs:
    for run in para.runs:
        for key, value in dict_modification.items():
            if key in run.text:
                run.text = run.text.replace(key, value)
                lista_de_substituicao = f"Substituído '{key}' por '{value}'"
                list_modification.append(lista_de_substituicao)
    print(para.text)

    # print("\n")
    # print(100 * '-')
    #print(f'{nome} - {df["CNPJ"].iloc[num]}')
# for item in list_modification:
#     print(item)

# Salva como DOCX
docx_path = r'../Arquivos/RECISÃO CONTRATUAL - HAP - NOTIFICAÇÃO COM AVISO - TESTE.docx'
pdf_path = r'../Arquivos/RECISÃO CONTRATUAL - HAP - NOTIFICAÇÃO COM AVISO - TESTE.pdf'
doc.save(docx_path)

# Converte para PDF
convert(docx_path, pdf_path)

# Remove o DOCX
os.remove(docx_path)
```
